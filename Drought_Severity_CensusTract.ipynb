```python
import geopandas
import matplotlib.pyplot as plt
# Link to information download: https://resilience-fema.hub.arcgis.com/datasets/FEMA::national-risk-index-annualized-frequency-drought/explore?layer=1&location=42.595305%2C-93.749460%2C7
gdf = geopandas.read_file("/content/Census_Tract.shp")
remove_states = ["Alaska","Hawaii", "Puerto Rico","Virgin Islands",
    "Guam","American Samoa", "Northern Mariana Islands"
]

# Creates new labeling 'keep' to keep all states that arent a part of the remove_states list
gdf["keep"] = ~gdf["STATE"].isin(remove_states)
usa = gdf[gdf["keep"]].copy()
usa.plot(
    column="DRGT_AFREQ",
    legend=True,
    figsize=(12, 8),
    cmap="YlOrRd",
    edgecolor="none",
    linewidth=0.1,
    legend_kwds={"label": "Drought Annualized Frequency"}
)

plt.axis("off")
plt.title("Drought Annualized Frequency by Census Tract")
plt.show()
```
