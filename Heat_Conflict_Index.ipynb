```python
import geopandas
import matplotlib.pyplot as plt
import os
# 2.32_August_UseMe
# Search up: EJScreen_2024_Tract_with_AS_CNMI_GU_VI.gdb

# Unzip file once -> then comment out
# !unzip /content/EJScreen_2024_Tract_with_AS_CNMI_GU_VI.gdb.zip

# One time use -> then comment out
# os.listdir("/content")

# Read file with geopandas
ej = geopandas.read_file("/content/EJScreen_2024_Tract_with_AS_CNMI_GU_VI.gdb")

# Get initial data
# print(ej.columns)
# ej.head()

# Remove these states -> make sure spelling is the same as in the data set using -> print(ej["STATE_NAME"].unique())
remove_states = ["ALASKA","HAWAII", "PUERTO RICO","Virgin Islands",
    "Guam","American Samoa", "Northern Mariana Is"]
ej["keep"] = ~ej["STATE_NAME"].isin(remove_states)
usa = ej[ej["keep"]].copy()

# Plot using updated information + data wanted
usa.plot(
    column="EXCEED_COUNT_80",
    legend=True,
    figsize=(14, 8),
    cmap="YlOrRd",
    edgecolor="none",
    linewidth=0.05,
    legend_kwds={"label": "EJ Indexes Over 80th Percentile"}
)
plt.axis("off")
plt.title("EPA Environmental Justice Screening Tool")
```
