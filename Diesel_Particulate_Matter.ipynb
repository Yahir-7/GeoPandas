```python
import geopandas
import matplotlib.pyplot as plt

# Link to data: https://public-environmental-data-partners.github.io/j40-cejst-2/en/downloads/
# Extract : "usa"

dpm = geopandas.read_file("/content/usa.shp")

# Initail infornation needed

#print(dpm["SF"].unique())
#print(dpm.columns)
#dpm.head()


remove_states = ["Alaska", "Hawaii", "Puerto Rico",
                 "American Samoa", "Guam", "Northern Mariana Islands",
                 "Virgin Islands"]

dpm["keep"] = ~dpm["SF"].isin(remove_states)
usa = dpm[dpm["keep"]].copy()

usa.plot(
    column = "DF_PFS",
    legend = True,
    figsize = (14, 8),
    cmap = "YlOrRd",
    linewidth = 0.05,
    edgecolor = "none",
    legend_kwds = {"label" : "Diesel PM Percentile"}
)

plt.axis("off")
plt.title("Diesel Particulate Matter by Census Tract")
plt.show()
```
